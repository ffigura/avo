### Plot all wells from Avseth et al. (2010)

```python
import numpy as np
import matplotlib.pyplot as plt
import las

%matplotlib inline
```

#### Import wells

```python
well_1 = las.LASReader('well_1.las', null_subs=np.nan)
well_2 = las.LASReader('well_2.las', null_subs=np.nan)
well_3 = las.LASReader('well_3.las', null_subs=np.nan)
well_4 = las.LASReader('well_4.las', null_subs=np.nan)
well_5 = las.LASReader('well_5.las', null_subs=np.nan)
```

#### Well 1

```python
log = ['GR', 'Vp', 'RHOB']
unit = ['API', 'km/s', 'g/cc']

f, ax = plt.subplots(nrows=1,ncols=3,figsize=(6,6))
for i in range(len(log)):
    ax[i].plot(well_1.data[log[i]],well_1.data['DEPT'],'-k')
    ax[i].set_title(log[i])
    ax[i].set_xlabel(unit[i])
    ax[i].invert_yaxis()
    ax[i].grid(linestyle='--')
        
plt.tight_layout()
```

#### Well 2

```python
log = ['GR', 'Vp', 'Vp','RHOB','NPHI']
unit = ['API', 'km/s','km/s','g/cc','%']
tops_well2={'Heimdal': 2153,'OWC': 2183}

f, ax = plt.subplots(nrows=1,ncols=5,figsize=(10,6))
for i in range(len(log)):
    ax[i].plot(well_2.data[log[i]],well_2.data['DEPT'],'-k')
    ax[i].set_title(log[i])
    ax[i].set_xlabel(unit[i])
    ax[i].invert_yaxis()
    ax[i].grid(linestyle='--')
    ax[i].axhline(y=float(tops_well2['Heimdal']),color='darkorange',lw=3,alpha=0.6)
    ax[i].axhline(y=float(tops_well2['OWC']),color='gray',lw=3,alpha=0.6)
    if i == 0:
        ax[i].text(3.3,tops_well2['Heimdal']-2,'Heimdal')
        ax[i].text(3.7,tops_well2['OWC']-2,'OWC')
    else:
        ax[i].set_yticklabels([])
        
plt.tight_layout()
```

#### Well 3

```python
log = ['GR', 'Vp', 'RHOB']
unit = ['API', 'km/s', 'g/cc']

f, ax = plt.subplots(nrows=1,ncols=3,figsize=(6,6))
for i in range(len(log)):
    ax[i].plot(well_3.data[log[i]],well_3.data['DEPT'],'-k')
    ax[i].set_title(log[i])
    ax[i].set_xlabel(unit[i])
    ax[i].invert_yaxis()
    ax[i].grid(linestyle='--')
        
plt.tight_layout()
```

#### Well 5

```python
log = ['GR', 'Vp', 'Vs', 'RHOB']
unit = ['API', 'km/s', 'km/s', 'g/cc']

f, ax = plt.subplots(nrows=1,ncols=4,figsize=(8,6))
for i in range(len(log)):
    if i == 1 or i == 2:
        ax[i].plot(well_5.data[log[i]]/100,well_5.data['DEPT'],'-k')
    else:
        ax[i].plot(well_5.data[log[i]],well_5.data['DEPT'],'-k')
    ax[i].set_title(log[i])
    ax[i].set_xlabel(unit[i])
    ax[i].invert_yaxis()
    ax[i].grid(linestyle='--')
        
plt.tight_layout()
```
